```python
from pandas import DataFrame
import pandas_profiling
import numpy as np
from matplotlib import pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
%matplotlib inline
```

```python
data = DataFrame.from_csv("./testosterone_levels.csv")
```

```python
pandas_profiling.ProfileReport(data)
```

```python
def get_sample_person(size=1):
    samples_ids = np.random.choice(data.index.unique(), size)
    return data.loc[samples_ids]
```

```python
def show_testosterone(personId):
    layout = dict(title = 'Variation of Tenstosterone Over Samples for Person ({})'.format(personId),
              xaxis = dict(title = 'Sample Number'),
              yaxis = dict(title = 'Testosterone'))

    observations = data[data.index.get_level_values(0) == index_to_pid[personId]]

    person = go.Scatter(x=np.arange(len(observations)), y=observations['Testosterone'], name='Person {}'.format(personId))
    
    iplot(dict(data = [person], layout = layout))
```

```python
index_to_pid = dict(list(zip(range(len(data.index.unique().values)), data.index.unique().values)))
```

```python
show_testosterone(49)
```

```python
sample_persons = get_sample_person(10)

layout = dict(title = 'Variation of Tenstosterone Over Samples for Each Person',
              xaxis = dict(title = 'Sample Number'),
              yaxis = dict(title = 'Testosterone'))

person_plot = []
for person_id, person_detail in sample_persons.groupby('Person'):
    person = go.Scatter(x=np.arange(len(sample_persons)), y=person_detail['Testosterone'], name='Person {}'.format(person_id))
    person_plot.append(person)

iplot(dict(data=person_plot, layout=layout))
```

```python
sample_person = get_sample_person().drop(["F1", "F2"], axis=1)
person_details = []
for feature in sample_person.columns:
    detail = go.Scatter(x= np.arange(len(sample_person)), y= sample_person[feature].values, name= feature)
    person_details.append(detail)
iplot(person_details)
```
